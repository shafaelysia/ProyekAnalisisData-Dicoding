## 3. main_df data exploration

```python
# Convert 'dteday' column to datetime format
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

# Merge day_df and hour_df
main_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="left",
    left_on="dteday",
    right_on="dteday"
)

# Check the first 5 rows
main_df.head()
```

```python
# Table overview
main_df.info()
```

# Data Visualization
## Visualisasi tren penyewaan sepeda secara keseluruhan

```python
# Aggregate the data by month
monthly_df = main_df.resample('M', on='dteday').sum()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(monthly_df.index, monthly_df['cnt_x'])
plt.xlabel('Month')
plt.ylabel('Number of Bike Rentals')
plt.title('Bike Rentals Over Time (Aggregated by Month)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi penyewaan sepeda berdasarkan musim, bulan, dan hari
### Berdasarkan Musim

```python
# Aggregate the data by season
season_agg = main_df.groupby("season_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = season_agg.index
y_max = season_agg[('cnt_x', 'max')]
y_min = season_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='lightblue')
plt.bar(x, y_min, label='Min Rentals', color='lightgreen')

# Modify x-axis labels
season_labels = ['1', '2', '3', '4']
plt.xticks(x, season_labels)

plt.xlabel('Season')
plt.ylabel('Number of Bike Rentals')
plt.title('Max and Min Bike Rentals by Season')
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i+1, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i+1, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')
    
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Berdasarkan Bulan

```python
# Aggregate the data by month
monthly_agg = main_df.groupby("mnth_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 7))
y = np.arange(len(monthly_agg.index))
y_max = monthly_agg[('cnt_x', 'max')]
y_min = monthly_agg[('cnt_x', 'min')]

bar_width = 0.5
plt.barh(y, y_max, height=bar_width, label='Max Rentals', color='lightblue')
plt.barh(y, y_min, height=bar_width, label='Min Rentals', color='lightgreen')

# Modify y-axis labels
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.yticks(y + bar_width / 2, month_labels)

plt.xlabel('Number of Bike Rentals')
plt.ylabel('Month')
plt.title('Max and Min Bike Rentals by Month')
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(max_val, i, str(max_val), ha='left', va='center', fontweight='bold')
    plt.text(min_val, i + bar_width, str(min_val), ha='left', va='center', fontweight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Berdasarkan Hari

```python
# Aggregate the data by weekday
weekday_agg = main_df.groupby("weekday_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = weekday_agg.index
y_max = weekday_agg[('cnt_x', 'max')]
y_min = weekday_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='lightblue')
plt.bar(x, y_min, label='Min Rentals', color='lightgreen')

plt.xlabel('Weekday')
plt.ylabel('Number of Bike Rentals')
plt.title('Max and Min Bike Rentals by Weekday')
plt.xticks(rotation=0)
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')
    
# Adjust the position of the legend
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi jam penyewaan sepeda paling banyak dan paling sedikit

```python
# Aggregate the data by hour
hourly_agg = main_df.groupby("hr").agg({
    "instant_y": "nunique",
    "cnt_y": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = hourly_agg.index
y_max = hourly_agg[('cnt_y', 'max')]
y_min = hourly_agg[('cnt_y', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='lightblue')
plt.bar(x, y_min, label='Min Rentals', color='lightgreen')

plt.xlabel('Hour')
plt.ylabel('Number of Bike Rentals')
plt.title('Max and Min Bike Rentals by Hour')

# Modify x-axis labels
hour_labels = [str(i) for i in x]
plt.xticks(x, hour_labels)

plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```
